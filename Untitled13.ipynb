```python
import re
import pandas as pd

pattern = re.compile(
    r'(?P<ip>\S+) \S+ \S+ '
    r'\[(?P<time>[^\]]+)\] '
    r'"(?P<method>\S+) (?P<url>\S+) \S+" '
    r'(?P<status>\d+) '
    r'\S+ '
    r'"(?P<referrer>[^"]*)" '
    r'"(?P<agent>[^"]*)"'
)

rows = []

with open("logfiles.log", "r", encoding="utf-8") as f:
    for line in f:
        match = pattern.search(line)
        if match:
            rows.append(match.groupdict())

df = pd.DataFrame(rows)
df.head()
```
